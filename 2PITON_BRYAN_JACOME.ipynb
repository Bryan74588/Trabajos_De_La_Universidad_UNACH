```python
# ===============================
# 1. IMPORTAR LIBRERÍAS
# ===============================
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ===============================
# 2. CARGAR DATASET
# ===============================
# Ajusta el separador (delimiter) a ';' para cargar correctamente el CSV
df = pd.read_csv("/content/d_2019.csv", sep=';')

# Vista rápida de las columnas
print(df.columns)
```

```python
df.dropna(inplace=True)
print(f"DataFrame 'df' ahora tiene {df.shape[0]} filas y {df.shape[1]} columnas después de eliminar filas con valores nulos.")
```

```python
map_sexo = {1:"Hombre", 2:"Mujer"}  # Hombre=1, Mujer=2
map_nac  = {1:"Ecuatoriano/a", 2:"Extranjero/a"}
map_cap  = {1:"Sí", 2:"No"}         # Capitulación: Sí=1, No=2
map_leer = {1:"Sí", 2:"No"}         # Sabe leer/escribir: Sí=1, No=2

map_inst = {
    0:"Ninguno", 1:"Alfabetización", 2:"Primaria", 3:"Educación básica",
    4:"Secundaria", 5:"Bachillerato", 6:"Superior no universitaria",
    7:"Superior universitaria", 8:"Posgrado"
}

map_causa = {
    1:"Mutuo cons. judicial",
    2:"Mutuo cons. notarial",
    3:"Adulterio",
    4:"Tratos crueles/violencia",
    5:"Falta de armonía habitual",
    6:"Amenazas graves",
    7:"Tentativa contra la vida",
    8:"Involucrar en ilícitos",
    9:"Condena > 10 años",
    10:"Ebriedad/toxicomanía",
    11:"Abandono > 6 meses"
}
# Autoidentificación étnica (INEC)
map_etnia = {
    1: "Indígena",
    2: "Afroecuatoriano/a",
    3: "Negro/a",
    4: "Mulato/a",
    5: "Montubio/a",
    6: "Mestizo/a",
    7: "Blanco/a",
    8: "Otro"
}

# Labels para etnia (Divorciado 1 y 2)
if "p_etnica_1" in df.columns:
    df["p_etnica_1_lbl"] = df["p_etnica_1"].map(map_etnia)

if "p_etnica2" in df.columns:
    df["p_etnica2_lbl"] = df["p_etnica2"].map(map_etnia)

# Labels para sexo (Divorciado 1 y 2)
if "sexo_1" in df.columns:
    df["sexo_1_lbl"] = df["sexo_1"].map(map_sexo)

if "sexo_2" in df.columns:
    df["sexo_2_lbl"] = df["sexo_2"].map(map_sexo)
```

```python
top_prov_insc = df["prov_insc"].value_counts().head(10)
top_prov_insc
```

```python
top_prov_insc.sort_values().plot(kind="barh")
plt.title("Top 10 provincias con más divorcios (prov_insc)")
plt.xlabel("Número de divorcios")
plt.ylabel("Provincia (código)")
plt.show()
```

```python
if "cau_div" in df.columns:
    df["cau_div_lbl"] = df["cau_div"].map(map_causa)
causa_rank = df["cau_div_lbl"].fillna(df["cau_div"].astype("Int64").astype(str)).value_counts()
causa_rank
causa_rank.plot(kind="bar")
plt.title("Causas de divorcio (INEC)")
plt.ylabel("Frecuencia")
plt.xticks(rotation=70, ha="right")
plt.show()

df["tipo_divorcio"] = np.where(df["cau_div"].isin([1,2]), "Mutuo consentimiento", "Causal (03-11)")
df["tipo_divorcio"].value_counts()
```

```python
sex_all = pd.concat([df["sexo_1_lbl"], df["sexo_2_lbl"]], ignore_index=True).dropna()
sex_all.value_counts()
sex_all.value_counts().plot(kind="bar")
plt.title("Distribución por sexo (divorciados 1 y 2)")
plt.ylabel("Personas")
plt.show()
```

```python
df["dur_mat"].describe()
df["dur_mat"].dropna().plot(kind="hist", bins=20)
plt.title("Distribución de duración del matrimonio (años)")
plt.xlabel("Años")
plt.ylabel("Frecuencia")
plt.show()
```

**Divorcios por nivel de Instrucción**

```python
# Mapear las columnas de nivel de instrucción a etiquetas descriptivas
df['niv_inst1_lbl'] = df['niv_inst1'].map(map_inst)
df['niv_inst2_lbl'] = df['niv_inst2'].map(map_inst)

# Combinar los niveles de instrucción de ambos divorciados en una sola serie
education_levels_all = pd.concat([df['niv_inst1_lbl'], df['niv_inst2_lbl']]).dropna()

# Contar la frecuencia de cada nivel de instrucción
education_counts = education_levels_all.value_counts()

# Mostrar la tabla de resultados
print("Distribución de divorcios por nivel de instrucción:")
display(education_counts)
```

```python
# Visualizar la distribución con un gráfico de barras
plt.figure(figsize=(10, 6))
education_counts.sort_values(ascending=True).plot(kind='barh')
plt.title('Divorcios por Nivel de Instrucción')
plt.xlabel('Número de Personas')
plt.ylabel('Nivel de Instrucción')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Divorcios por Etnia**

```python
# Combinar las etiquetas de etnia de ambos divorciados en una sola serie
etnia_levels_all = pd.concat([df['p_etnica_1_lbl'], df['p_etnica2_lbl']]).dropna()

# Contar la frecuencia de cada nivel de etnia
etnia_counts = etnia_levels_all.value_counts()

# Mostrar la tabla de resultados
print("Distribución de divorcios por Etnia:")
display(etnia_counts)

# Visualizar la distribución con un gráfico de barras
plt.figure(figsize=(10, 6))
etnia_counts.sort_values(ascending=True).plot(kind='barh')
plt.title('Divorcios por Etnia')
plt.xlabel('Número de Personas')
plt.ylabel('Etnia')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**En que mes exstieron mas divorcios**

```python
# Crear un mapeo de números de mes a nombres de mes
map_mes = {
    1: "Enero", 2: "Febrero", 3: "Marzo", 4: "Abril",
    5: "Mayo", 6: "Junio", 7: "Julio", 8: "Agosto",
    9: "Septiembre", 10: "Octubre", 11: "Noviembre", 12: "Diciembre"
}

# Mapear la columna 'mes_insc' a los nombres de los meses
df['mes_insc_lbl'] = df['mes_insc'].map(map_mes)

# Contar la frecuencia de divorcios por mes
divorcios_por_mes = df['mes_insc_lbl'].value_counts().sort_index()

print("Divorcios por Mes:")
display(divorcios_por_mes)

# Identificar el mes con más divorcios
mes_mas_divorcios = divorcios_por_mes.idxmax()
num_mas_divorcios = divorcios_por_mes.max()
print(f"\nEl mes con más divorcios fue: {mes_mas_divorcios} con {num_mas_divorcios} divorcios.")
```

```python
# Visualizar la distribución de divorcios por mes
plt.figure(figsize=(12, 6))
divorcios_por_mes.plot(kind='bar', color='skyblue')
plt.title('Divorcios por Mes de Inscripción')
plt.xlabel('Mes')
plt.ylabel('Número de Divorcios')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Cuál es la media mediana y moda de la duración de matrimonios, antes del divorcio**

```python
# Calcular la media de la duración del matrimonio
mean_dur_mat = df['dur_mat'].mean()

# Calcular la mediana de la duración del matrimonio
median_dur_mat = df['dur_mat'].median()

# Calcular la moda de la duración del matrimonio
mode_dur_mat = df['dur_mat'].mode()

print(f"Media de la duración del matrimonio: {mean_dur_mat:.2f} años")
print(f"Mediana de la duración del matrimonio: {median_dur_mat:.2f} años")
print(f"Moda de la duración del matrimonio:\n{mode_dur_mat.to_string(index=False)} años")
```

**Cuál es la mayor edad y la menor en la que se ha registrado un divorcio**

```python
# Combinar las columnas de edad de ambos divorciados
ages_all = pd.concat([df['edad_1'], df['edad_2']]).dropna()

# Convertir a tipo entero (si no lo está ya) y manejar posibles valores no numéricos
ages_all = pd.to_numeric(ages_all, errors='coerce').dropna().astype(int)

# Calcular la edad mínima
min_age = ages_all.min()

# Calcular la edad máxima
max_age = ages_all.max()

print(f"La edad mínima registrada en un divorcio es: {min_age} años")
print(f"La edad máxima registrada en un divorcio es: {max_age} años")
```

**Crear un gráfico de cuántas personas divorciadas que saben leer y que no saben leer**

```python
# Mapear las columnas 'sabe_leer' a etiquetas descriptivas
df['sabe_leer1_lbl'] = df['sabe_leer1'].map(map_leer)
df['sabe_leer2_lbl'] = df['sabe_leer2'].map(map_leer)

# Combinar los datos de ambas columnas en una sola serie
sabe_leer_all = pd.concat([df['sabe_leer1_lbl'], df['sabe_leer2_lbl']]).dropna()

# Contar la frecuencia de 'Sí' y 'No'
sabe_leer_counts = sabe_leer_all.value_counts()

print("Distribución de personas divorciadas por habilidad para leer:")
display(sabe_leer_counts)

# Visualizar la distribución con un gráfico de barras
plt.figure(figsize=(7, 5))
sabe_leer_counts.plot(kind='bar', color=['lightcoral', 'skyblue'])
plt.title('Personas Divorciadas: ¿Saben Leer?')
plt.xlabel('Habilidad para Leer')
plt.ylabel('Número de Personas')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
