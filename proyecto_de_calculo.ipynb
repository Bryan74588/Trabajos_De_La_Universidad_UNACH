```python
# ============================================
# C√ÅLCULO DE √ÅREAS DE TERRENOS IRREGULARES
# M√âTODO DEL TRAPECIO
# TOPOGRAF√çA E INGENIER√çA CIVIL
# ============================================

import matplotlib.pyplot as plt

def area_trapecio(ordenadas, distancia):
    """
    Calcula el √°rea usando el m√©todo del trapecio
    """
    n = len(ordenadas)

    if n < 2:
        return 0

    area = (distancia / 2) * (ordenadas[0] + ordenadas[-1] + 2 * sum(ordenadas[1:-1]))
    return area


# ---------- INGRESO DE DATOS ----------
n = int(input("Ingrese el n√∫mero de ordenadas: "))
distancia = float(input("Ingrese la distancia entre secciones (m): "))

ordenadas = []
for i in range(n):
    y = float(input(f"Ingrese la ordenada {i+1}: "))
    ordenadas.append(y)


# ---------- C√ÅLCULO DEL √ÅREA ----------
area = area_trapecio(ordenadas, distancia)


# ---------- COORDENADAS PARA LA GR√ÅFICA ----------
x = [i * distancia for i in range(n)]


# ---------- GR√ÅFICA ----------
plt.figure(figsize=(10, 5))

plt.plot(x, ordenadas, marker='o', linewidth=2)
plt.fill_between(x, ordenadas, alpha=0.3)

plt.xlabel("Distancia (m)")
plt.ylabel("Ordenadas (m)")
plt.title("C√°lculo de √Årea por el M√©todo del Trapecio")
plt.grid(True)

plt.text(
    0.05 * max(x),
    0.9 * max(ordenadas),
    f"√Årea = {area:.2f} m¬≤",
    fontsize=12,
    bbox=dict(facecolor='white')
)

plt.show()


# ---------- RESULTADO ----------
print(f"\n√Årea total del terreno = {area:.2f} m¬≤")
```
