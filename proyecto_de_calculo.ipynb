```python
# ---------------------------------------------------------
# PROGRAMA PARA CALCULAR:
# 1) √ÅREA DE UN TERRENO IRREGULAR
# 2) LONGITUD DE LA CURVA DEL TERRENO (PERFIL)
# USANDO EL M√âTODO DEL TRAPECIO
# CON GR√ÅFICA
# ---------------------------------------------------------

# Importamos la librer√≠a para hacer gr√°ficas
import matplotlib.pyplot as plt

# Importamos la librer√≠a matem√°tica
import math


# -------- FUNCI√ìN PARA CALCULAR EL √ÅREA --------
def area_trapecio(ordenadas, distancia):
    """
    Calcula el √°rea usando el m√©todo del trapecio
    """

    # Cantidad de ordenadas
    n = len(ordenadas)

    # Si no hay suficientes datos, no hay √°rea
    if n < 2:
        return 0

    # F√≥rmula del m√©todo del trapecio
    area = (distancia / 2) * (
        ordenadas[0] +            # Primera ordenada
        ordenadas[-1] +           # √öltima ordenada
        2 * sum(ordenadas[1:-1])  # Ordenadas intermedias por 2
    )

    # Retornamos el √°rea
    return area


# -------- FUNCI√ìN PARA CALCULAR LA CURVA --------
def longitud_curva(ordenadas, distancia):
    """
    Calcula la longitud real de la curva del terreno
    """

    # Inicializamos la longitud total en 0
    longitud = 0

    # Recorremos cada tramo del perfil
    for i in range(len(ordenadas) - 1):

        # Diferencia horizontal (dx)
        dx = distancia

        # Diferencia vertical (dy)
        dy = ordenadas[i+1] - ordenadas[i]

        # F√≥rmula de distancia entre dos puntos:
        # L = sqrt(dx¬≤ + dy¬≤)
        tramo = math.sqrt(dx**2 + dy**2)

        # Sumamos cada tramo a la longitud total
        longitud += tramo

    # Retornamos la longitud total de la curva
    return longitud


# -------- INGRESO DE DATOS --------

# Pedimos el n√∫mero de ordenadas
n = int(input("Ingrese el n√∫mero de ordenadas: "))

# Pedimos la distancia entre secciones
distancia = float(input("Ingrese la distancia entre secciones (m): "))

# Creamos la lista de ordenadas
ordenadas = []

# Ingreso de cada ordenada
for i in range(n):
    y = float(input(f"Ingrese la ordenada {i+1}: "))
    ordenadas.append(y)


# -------- C√ÅLCULOS --------

# Calculamos el √°rea
area = area_trapecio(ordenadas, distancia)

# Calculamos la longitud de la curva
curva = longitud_curva(ordenadas, distancia)


# -------- COORDENADAS PARA LA GR√ÅFICA --------

# Creamos las distancias acumuladas en X
x = [i * distancia for i in range(n)]


# -------- GR√ÅFICA --------

# Tama√±o de la gr√°fica
plt.figure(figsize=(10, 5))

# Dibujamos el perfil del terreno
plt.plot(x, ordenadas, marker='o', linewidth=2)

# Rellenamos el √°rea del terreno
plt.fill_between(x, ordenadas, alpha=0.3)

# Etiquetas
plt.xlabel("Distancia (m)")
plt.ylabel("Ordenadas (m)")
plt.title("√Årea y Curva del Terreno - M√©todo del Trapecio")

# Cuadr√≠cula
plt.grid(True)

# Texto con resultados en la gr√°fica
plt.text(
    0.05 * max(x),
    0.85 * max(ordenadas),
    f"√Årea = {area:.2f} m¬≤\nCurva = {curva:.2f} m",
    fontsize=12,
    bbox=dict(facecolor='white')
)

# Mostramos la gr√°fica
plt.show()


# -------- RESULTADOS --------

print("\n====================================")
print(f"√Årea total del terreno = {area:.2f} m¬≤")
print(f"Longitud de la curva = {curva:.2f} m")
print("====================================")

```
